Mizzare Corporation has collected data on customer satisfaction levels for two types of smart home devices: Smart Thermostats and Smart Lights. They want to determine if there's a significant association between the type of device purchased and the customer's satisfaction level.
Data Provided:

Satisfaction	Smart Thermostat	Smart Light	Total

```python
#H0: There is no association between Device Type and Customer Satisfaction.
#H1: There is an association between Device Type and Customer Satisfaction.
```

```python
df=pd.DataFrame({'Satisfaction':['Very Satisfied','Satisfied','Neutral','Unsatisfied','Very Unsatisfied'],'Smart Thermostat':[50,80,60,30,20],'Smart Light':[70,100,90,50,50]})
df

```

```python
df.set_index('Satisfaction',inplace=True)
df
```

```python
stats.chi2_contingency(df)
```

```python
chi_critical=stats.chi2.ppf(0.95,4)
chi_critical
```

```python
#chi.stats<chi_critical then fail to reject null hypothesis.
#Then we conclude that there is no association between Device Type and Customer Satisfaction.
```

```python
##We are testing whether the sample mean weekly cost is significantly different from the theoretical mean.

Null Hypothesis (H₀): μ = 4000
(The mean weekly cost is Rs. 4,000, as per the model.)

Alternative Hypothesis (H₁): μ ≠ 4000
(The mean weekly cost is not equal to Rs. 4,000.)
```

```python
# from given data
x_bar=3050
n=25
pop_mean=1000+5*(600)
sigma=25*5
```

```python
z_score=(x_bar-pop_mean)/(sigma/np.sqrt(n))
z_score
```

```python
z_critical=stats.norm.ppf(0.975)
z_critical
```

```python
# a_critical value=+1.95 to -1.95
#since z_score<z_critical
# then we reject null hypothesis
#There is strong evidence to suggest that the actual mean weekly cost differs significantly from the theoretical cost predicted by the model.
```
