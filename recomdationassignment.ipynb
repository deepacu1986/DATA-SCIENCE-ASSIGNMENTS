```python
anime_df['rating']=anime_df['rating'].fillna(anime_df['rating'].mean())
anime_df['type']=anime_df['type'].fillna(anime_df['type'].mode()[0])
anime_df['genre']=anime_df['genre'].fillna(anime_df['genre'].mode()[0])
```

```python
# One-hot encode genres
from sklearn.preprocessing import MultiLabelBinarizer

# Split genres and binarize
# The genre column is already split into lists, so no need to split again.
mlb = MultiLabelBinarizer()
genre_matrix = mlb.fit_transform(anime_df['genre'])
# Normalize episodes and rating
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
anime_df[['rating']] = scaler.fit_transform(anime_df[['rating']])


# Create DataFrame of genres
genre_df = pd.DataFrame(genre_matrix, columns=mlb.classes_)
df1=anime_df.pivot_table(index='name',values='rating')
df1.fillna(0,axis=1,inplace=True)
df1


```

```python
from sklearn.metrics.pairwise import cosine_similarity

# Compute cosine similarity matrix
cos_sim = cosine_similarity(df1)
# Map indices to anime titles
anime_indices = pd.Series(anime_df.index, index=anime_df['name']).drop_duplicates()
anime_indices.head()
```

```python
# item based similarity
```

```python
def recommended_movie(similar_movie):
  if similar_movie in anime_indices.index:
    index=anime_indices.index.get_loc(similar_movie)
    similar_movies=sorted(list(enumerate(cos_sim[index])),reverse=True,key=lambda x:x[1])[1:6]
    print(f'Recommaned movie of {similar_movie} are:')
    print('*'*27)
    for movie in similar_movies:
      print(anime_indices.index[movie[0]])
  else:
    print(f'{similar_movie} is not in our database')
```
